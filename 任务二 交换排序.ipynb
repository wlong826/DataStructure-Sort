## 1. 冒泡排序

　　冒泡排序（Bubble Sort）是一种最简单的交换排序方法，它通过两两比较相邻记录的关键字，如果发生逆序，则进行交换，从而使关键字小的记录如气泡一般逐渐往上"漂浮"(左移)，或者使关键字大的记录如石块一样逐渐向下"坠落"(右移)。

### 1.1 冒泡排序算法函数

```python
def BubbleSort(sqlist):
    m = len(sqlist)-1
    flag = 1 # 用来标记
    while(m > 0 and flag == 1):
        flag = 0 # flag置为0，如果本趟排序没有发生交换，则不会执行下一趟排序
        for i in range(m):
            if sqlist[i] > sqlist[i+1]:
                flag = 1 # flag置为1表示，本趟排序发生了交换
                t = sqlist[i] # 交换前后两个记录
                sqlist[i] = sqlist[i+1]
                sqlist[i+1] = t
        m = m - 1
    return sqlist
```

### 1.2 测试

```python
sqlist = [49, 38, 65, 97, 76, 13, 27, 49, 55, 4]
print(" 原始序列 ：", sqlist)
print("排序后序列：", BubbleSort(sqlist))
```

## 2. 快速排序

　　快速排序（Quick Sort）是由冒泡排序改进而得的。在冒泡排序过程中，只对相邻的两个记录进行比较，因此每次交换两个相邻记录时只能消除一个逆序。如果能通过两个（不相邻）记录的一次交换，消除多个逆序，则会大大加快排序的速度。快速排序方法中的一次交换可能消除多个逆序。

### 2.1 快速排序算法函数

```python
## 对顺序表sqlist中的子表进行一次排序，返回枢轴位置
def Partition(sqlist, low, high):
    temp = sqlist[low] # 记录枢轴记录
    pivotkey = sqlist[low] # 枢轴记录关键字保存在pivotkey
    while low < high:
        while low < high and sqlist[high] >= pivotkey:
            high = high - 1
        sqlist[low] = sqlist[high] # 将比枢轴记录小的记录移到低端
        while low < high and sqlist[low] <= pivotkey:
            low = low + 1
        sqlist[high] = sqlist[low] # 将比枢轴记录大的记录移到高端
    sqlist[low] = temp # 枢轴记录到位
    return low

## 采用递归对顺序表进行快速排序
def QSort(sqlist, low, high):
    if low < high: # 长度大于1 
        pivotloc = Partition(sqlist, low, high) # 将[low, high]一分为二，pivotloc是枢轴位置
        QSort(sqlist, low, pivotloc-1) # 对左子表递归排序
        QSort(sqlist, pivotloc+1, high) # 对右子表递归排序
        
## 对顺序表快速排序
def QuickSort(sqlist):
    low = 0
    high = len(sqlist) - 1
    QSort(sqlist, 0, high)
    return sqlist
```

### 2.2 测试

```python
sqlist = [49, 38, 65, 97, 76, 13, 27, 49]
print(" 原始序列 ：", sqlist)
print("排序后序列：", QuickSort(sqlist))
```
