**算法基本步骤：**

　（1）归并排序（Merging Sort）（难度：★★）
 
　　基本思想：假设初始序列含有n个记录，则可看成是 n个有序的子序列，每个子序列的长度为1，然后两两归并，得到[n/2]个长度为2或1的有序子序列；再两两归并，……， 如此重复，直至得到一个长度为n 的有序序列为止。
